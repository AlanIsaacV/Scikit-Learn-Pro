```python
from IPython.display import display

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import MeanShift

%matplotlib inline
```

```python
df = pd.read_csv('../data/candy.csv')
x = df.drop(columns=['competitorname']).values

df.head(2)
```

```python
meanshift = MeanShift()
df['cluster'] = meanshift.fit_predict(x)
```

```python
sns.pairplot(df, hue='cluster', palette='Pastel1', plot_kws=dict(s=150, edgecolor="w", linewidth=1))
```
